```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('owid-covid-data (1).csv')
df.head()
```

## Continent

```python
df_continent = df.copy()
df_continent = df_continent['continent'].unique()
df_continent = pd.DataFrame(df_continent)
df_continent.to_csv('continent.csv', header=True)
df_continent
```

## Country

```python

temp = list(df_continent[0])
def set_key(countryName):
    return temp.index(countryName) + 1

df_country = df.copy()
df_country = df_country[['location', 'continent']]
df_country = df_country.drop_duplicates(subset=None, keep='first', inplace=False)
# df_country = pd.DataFrame(df_country)
df_country['con-id'] = np.vectorize(set_key)(df_country['continent'])
df_country.to_csv('country.csv', header=True)
df_country.head()
```

## TotalCases

```python
df_cases = df.copy()
df_cty_id = df_country.copy()
df_cty_id = df_cty_id.reset_index().reset_index().drop('index', axis=1)
cty_id = list(df_cty_id['location'])
def set_key(countryName):
    return cty_id.index(countryName) + 1

df_cases = df_cases[['total_cases', 'date', 'location']]
df_cases['cty_id'] = np.vectorize(set_key)(df_cases['location'])
df_cases = df_cases
df_cases.to_csv('totalcases.csv', header=True)
df_cases.head()
```

## New Cases

```python
df_new_cases = df.copy()
df_cty_id = df_country.copy()
df_cty_id = df_cty_id.reset_index().reset_index().drop('index', axis=1)
cty_id = list(df_cty_id['location'])
def set_key(countryName):
    return cty_id.index(countryName) + 1

df_new_cases = df_new_cases[['new_cases', 'new_cases_smoothed', 'date', 'location']]
df_new_cases['cty_id'] = np.vectorize(set_key)(df_new_cases['location'])
df_new_cases = df_new_cases.fillna(0)
df_new_cases.to_csv('newcases.csv', header=True)
df_new_cases.head()
```
